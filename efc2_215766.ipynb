```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

#definindo N e k
N = 200
k1 = 1
k2 = 10
k3 = 20
k4 = 50

#parametros de x e y
start = -4
stop = 4
mean = 1
sigma = 2

#definindo x e y periodo e w0
x = np.linspace(start,stop,N,endpoint=True)
y = np.random.normal(mean, sigma, len(x))
periodo = 4
w0 = 2*np.pi/periodo

#define coeficiente ak
def ak(k):
    if k == 0:
      a = 1
    else:
      a = (np.sin(2*k*w0))/(2*k*w0) + 1j*(1/(k*w0))*((np.sin(2*k*w0))/(2*k*w0)-(np.cos(2*k*w0)))
    return a.sum()/a.size

#define soma de fourier
def f(x, Nh):
    f = np.array([2*ak(i)*np.exp(1j*i*w0*x) for i in range(1,Nh+1)])
    return f.sum()


#aplica soma de fourier para diferentes Ns

Fourier1 = np.array([f(t,k1).real for t in x])
Fourier2 = np.array([f(t,k2).real for t in x])
Fourier3 = np.array([f(t,k3).real for t in x])
Fourier4 = np.array([f(t,k4).real for t in x])

#plota
fig = plt.figure(figsize=(8, 6), dpi=80,facecolor='yellow')
ax = fig.add_subplot(111)
plt.title("SÃ©rie de Fourier para uma onda Dente de Serra")
ax.plot(x , Fourier1 +1 , label = (" k = %d " % k1))
ax.plot(x , Fourier2 +1 , label = (" k = %d " % k2))
ax.plot(x , Fourier3 +1 , label = (" k = %d " % k3))
ax.plot(x , Fourier4 +1 , label = (" k = %d " % k4))
ax.plot(x, 1 -2*signal.sawtooth(w0*(x-2) ))
ax.grid(True, color='dimgray', linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)
ax.legend()
ax.set_ylabel('x(t)')
ax.set_xlabel('t')
```
